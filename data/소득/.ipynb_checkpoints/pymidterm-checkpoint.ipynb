```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname = "C:\\Windows\\Fonts\\malgun.ttf").get_name()
plt.rc("font", family=font_name)
```

```python
earnings = pd.read_csv("data/2021earnings.csv")

earnings
```

```python
satisfaction = pd.read_csv("data/satisfaction.csv" ,encoding = 'cp949')
satisfaction
```

```python
# 소득데이터와 만족도데이터 병합

bind = pd.merge(earnings, satisfaction, left_on = '구', right_on = 'gu_name')

del bind["Gu"]
del bind["gu_name"]

bind
```

```python
# 소득 범주의 비율 칼럼 추가

bind['200미만비'] = round(bind['200미만']/(bind['200미만']+bind['200이상500미만']+bind['500이상'])*100, 2)
bind['200이상500미만비'] = round(bind['200이상500미만']/(bind['200미만']+bind['200이상500미만']+bind['500이상'])*100 ,2)
bind['500이상비'] = round(bind['500이상']/(bind['200미만']+bind['200이상500미만']+bind['500이상'])*100 ,2)

bind
```

```python
# 체크포인트

bind.to_csv("data/bind.csv")
```

### 지역 별 소득범주 비율 시각화

```python
# 지역 별 소득 범주 시각화

bind2 = bind[['구','200미만비','200이상500미만비','500이상비']]
bind2 = bind2.set_index(['구'])
bind2

bind2.plot(kind="bar", stacked=True)
```

```python
# 소득 높고 낮은 지역을 고를 때 200이상500미만비 범주는 보지 않으므로 제외하고 시각화

bind3 = bind[['구','200미만비','500이상비']]
bind3 = bind3.set_index(['구'])

bind3.plot(kind="barh", stacked = True)
```

### 소득 상위지역, 하위지역 시각화

```python
# 한달 소득 200미만 상위5개, 하위 5개 지역 시각화 (비율)


class1 = bind[['구','200미만비']]
class1
class1_1 = class1.sort_values('200미만비')
class1_1 = class1_1.head(5).append(class1_1.tail(5))
class1_1
plt.pie(class1_1['200미만비'], labels = class1_1['구'])
```

```python
# 한달 소득 500이상 상위5개, 하위 5개 지역 시각화 (비율)

class3 = bind[['구','500이상비']]
class3
class3_1 = class3.sort_values('500이상비')
class3_1 = class3_1.head(5).append(class3_1.tail(5))
class3_1
plt.pie(class3_1['500이상비'], labels = class3_1['구'])
```


저소득 구간의 비율이 높은 지역일수록...  : 이 때, 저소득 구간의 비율 = 월소득 200미만 구간의 비율

고소득 구간의 비율이 높은 지역일수록...  : 이 때, 고소득 구간의 비율 = 월소득 500이상 구간의 비율



여기까지 하구 r로 상관계수 확인했습니다

### R을 통한 상관분석 결과

#### 저소득 구간 비율과 삶만족도의 상관계수는 -0.21로 약한 음의 상관관계가 있다고 볼 수 있다.
#### 저소득 구간 비율과 지역만족도의 상관계수는 -0.5로 음의 상관관계가 있다고 볼 수 있다.
#### 고소득 구간 비율과 삶만족도의 상관계수는 0.15로 아주 약한 양의 상관관계가 있다고 볼 수 있다.
#### 저소득 구간 비율과 지역만족도의 상관계수는 0.42로 양의 상관관계가 있다고 볼 수 있다. 

### 상관관계 시각화

```python

bind3 = bind[['구','삶만족도','지역만족도','200미만비','200이상500미만비','500이상비']]
bind3 = bind3.set_index(['구'])
bind3
```

```python
# 월소득 200미만 비율과 삶만족도의 관계 시각화
sns.lmplot(x="200미만비", y="삶만족도", data=bind3)       # fit_reg=False     회귀선 숨긴버전
```

```python
# 월소득 200미만 비율과 지역만족도의 관계 시각화
sns.lmplot(x="200미만비", y="지역만족도", data=bind3)
```

```python
# 월소득 500이상 비율과 삶만족도의 관계 시각화
sns.lmplot(x="500이상비", y="삶만족도", data=bind3)
```

```python
# 월소득 500이상 비율과 지역만족도의 관계 시각화
sns.lmplot(x="500이상비", y="지역만족도", data=bind3)
```
